###### Applying Detection and Containment Processes
***

Having the ability to make tactical decisions on the fly is important as you work your way through the incident response lifecycle. 

> ``OODA Loop``

Can help practitioners make these decisions in real time by using these 4 steps:

- Observe

Ensure you have the correct data and a way of searching through it.

- Orient

Establish where you are currently in the attack process.

- Decide

Decide on what the overall goal is, including countermeasures or prevention techniques.

- Act

Act on the decisions of the first three stages and repeat until intended outcome is achieved.

>  Course of Action Matrix - ``CoA``

Another applied method to help the decision making process during an Incident is used by the Department of Defense and called the Course of Action (``CoA``) matrix. 

It is as follows:

- Detect

Identify the presence of a threat.

- Destroy

Render the threat's attack useless.

- Degrade

Reduce the threat by limiting capabilities.

- Disrupt

Attempt to attack back by using tactics like confusion.

- Deny

Block all attempts by the threat.

- Deceive

Provide false information to buy time or deter the adversary.

###### Impact Analysis
***

Being able to quickly determine the impact on the Company affected is essential in determining next steps that need to be taken. 

The following are some ways to differentiate impact at the organizational level:

- Organizational Impact

Affects wide-scale, mission essential functions that could affect operability.

- Localized Impact

Small instances, isolated to groups or workstations.

- Immediate Impact

Direct costs affected including fee, penalties, and downtime.

- Total Impact

Costs that follow the incident like damage to company reputation.

###### Containment
***

Once you have determined that a machine is compromised, the following steps should be taken to quickly contain the host for further analysis:

- Ensure safety of personnel and security first and foremost.


- Prevent ongoing or further contamination.


- Identify the primary intrusion.


- Try to avoid alerting the attacker.


- Keep a good forensic image of the situation.

Containment can be accomplished in one of two ways:

1. Isolation-based Containment 

Is the most effective way to stop the spread of further contamination of the attack, but may limit your visibility into the host for the time being.

2. Segmentation-based Containment 

Allows you to monitor the assets' activities in a controlled environment to further try and identify the threat actor and their motives.
