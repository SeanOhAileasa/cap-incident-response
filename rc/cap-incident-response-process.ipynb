###### Incident Response Phases
***

Incident response procedures are the actions and guidelines for dealing with a security event.

[ [``NIST``](https://nvlpubs.nist.gov/nistpubs/SpecialPublications/NIST.SP.800-61r2.pdf) ] publishes a special for security incident handling to help guide response efforts from preparation to post incident - below shows the phases in this lifecycle:

Below is a breakdown of what these phases represent.

> Preparation

Steps here include system hardening, establishing policies and secure lines of communication.

> Detection and Analysis

Having the ability to quickly determine if an incident has occured.

> Containment

Restrict or limit the total damage that could be done in the case of an incident.

> Eradication and Recovery

The ability to remove or patch the threat, then bring the affected system back online and into a normal state.

> Post Incident Activity

Continuous improvement and review to prevent future instances from occurring.

***
###### Documentation of Procedures
***

It is vital that all procedures are well documented and can be clearly understood as to how they should be executed. 

All response procedures should be contained inside a centralized incident response plan or ``IRP``. 

- Plan will contain all playbooks of how each type of event should be executed.

Should contain your call list of all important contacts that may be needed once an incident has been officially documented. 

- Within this call list, should also contain an escalation list of who should be contacted first and so on.

###### Incident Form
***

Once an incident has been discovered and confirmed, it will need to be officially documented on an incident form. 

This form should contain the following features:

- Location of incident


- Date/time of incident


- Name of incident reporter and all other parties involved


- How incident was discovered


- Incident type (malware, data breach, etc.)


- Scope of affected systems


- Incident event timeline

###### Data Criticality and Prioritization
***

When dealing with an incident or preparing an Incident Response Plan (``IRP``), it is important to understand the different types of data you are dealing with and their criticality. 

The following are data labels that are used to classify different types of data:

> Personally Identifiable Information (``PII``)

Data that can be used to identify an individual (or in the case of identity theft, impersonate them).

> Sensitive Personal Information (``SPI``)

Information about a subject's opinions, beliefs, and nature that is afforded specially protected status by privacy legislation.

> Personal Health Information (``PHI``)

Data that can be used to identify someone as the subject of medical and insurance records, plus associated hospital and lab test results.

> Financial Information

Data held about bank and investment accounts, plus information about payroll and tax returns.

> Intellectual Property

Information created by a company, typically about the products or services that they make or perform.

> Corporate Information

Confidential data owned by a company, such as product, sales, marketing, and legal information.

> High Value Assets

An information asset that processes data critical to a mission essential function.

###### Communication Plan
***

Having an “out-of-band” communication plan in place is a critical control to have in place in the scenario that you are unable to execute your main method of communication - solutions:

- Include a messaging system that includes end-to-end encryption.


- Outside email system that has message encryption enabled like ``S/MIME``.

###### Reporting Requirements
***

Reporting requirements are notifications that must be made to affected parties in the event of a data breach, as required by legislation or regulation entities.

One of the most important requirements in the ``IRP`` is reporting the type of incident involved - will determine which procedures need to be followed moving forward. 

The following are some common types of incident reporting:

- Accidental Breach

Either a misconfiguration or a human mistake causes the breach of data.

- Data Exfiltration or Insider Data Exfiltration

Where classified data is transferred to another system or outside of your controlled network by an external party or even an internal party.

- Device Loss/Theft

Loss of a device storing sensitive data.

- Integrity or Availability of Data

###### Response Coordination
***

When establishing your call escalation list, will need to have an understanding of who needs to be involved and contacted at specific times during the ``IRP`` process.

The following are some common groups or individuals that will be involved in the ``IRP`` process:

- Senior Leadership


- Regulatory Bodies


- Law Enforcement


- Legal Team


- Public Relations


- Human Resources

***
###### IR Training and Testing
***

Last but not least, once you have an ``IRP`` in place, it will need to be practiced, and all involved parties trained on response procedures on a recurring basis.

> Non-technical Roles

Table-top exercises should be executed to formally walk through different types of scenarios to ensure all parties are aware of their responsibilities and what they need to do in each type of situation.

> Technical Roles

Penetration testing exercises can be performed to help hone in on response times and detection capabilities, as well as which steps to take to eradicate potential threats in a controlled environment.
